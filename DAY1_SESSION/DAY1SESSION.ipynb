### READ THE FILE

```python
df = spark.read.csv("/Volumes/workspace/default/capgemini_training/C1.csv",header=True,inferSchema=True)
```

```python
df.display()
```

### Handling 'blank'/'Blank' as NULL

```python
from pyspark.sql.functions import *
df = df.replace(["blank","Blank"],None)
```

```python
df.display()
```

### Fix Country Column india / INDIA / USA -> inconsistent

```python
df = df.withColumn("Country", upper(col("Country")))
```

```python
df.display()
```

### Fix DATE Format (15-01-2023,2023)

### Fill Values

```python
df=df.fillna({
    "Category":"Unknown",
    "Sales": 0
})
```

```python
df.display()
```

### Remove Duplicates

```python
df=df.dropDuplicates((["Name"]))
df.display()
```
