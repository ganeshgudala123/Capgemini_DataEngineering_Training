### SAMPLE TABLES

```sql
drop table Employees;
drop table Departments;
drop table Orders;
```

```sql
CREATE TABLE Employees (
    emp_id INT PRIMARY KEY,
    name VARCHAR(50),
    department_id INT,
    salary INT,
    manager_id INT
);
CREATE TABLE Departments (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(50)
);
CREATE TABLE Orders (
    order_id INT PRIMARY KEY,
    emp_id INT,
    order_amount INT,
    order_date DATE
);
INSERT INTO Employees VALUES
(1, 'Alice', 101, 60000, NULL),
(2, 'Bob', 102, 45000, 1),
(3, 'Charlie', 101, 70000, 1),
(4, 'David', 103, 40000, 2),
(5, 'Eve', 102, 50000, 2);
INSERT INTO Departments VALUES
(101, 'HR'),
(102, 'IT'),
(103, 'Finance');
INSERT INTO Orders VALUES
(1, 1, 5000, '2024-01-10'),
(2, 2, 3000, '2024-02-15'),
(3, 3, 7000, '2024-03-12'),
(4, 1, 2000, '2024-04-01'),
(5, 4, 1000, '2024-05-05');
```

### SUB QUERIES TOPIC - Find employees who earn more than average salary

```sql
SELECT emp_id
FROM Employees
WHERE salary > (
    SELECT AVG(salary)
    FROM Employees
);
```

### Find employees who belong to the same department as 'Alice'

```sql
SELECT *
FROM Employees
WHERE department_id = (
    SELECT department_id
    FROM Employees
    WHERE name = 'Alice'
);
```

### Get employees whose salary is equal to the minimum salary

```sql
select*from Employees where salary = (select MIN(salary)from Employees)
```

### Find employees who have placed at least one order

```sql
SELECT emp_id, name
FROM Employees
WHERE emp_id IN (
    SELECT emp_id
    FROM Orders
);
```

### CTE (Common Table Expression)

```sql
--Create a CTE to show employees with salary > 50,000 and fetch all records
WITH HighSalaryEmployees AS (
    SELECT *
    FROM Employees
    WHERE salary > 50000
)
SELECT *
FROM HighSalaryEmployees;
```

```sql
--Find average salary per department using CTE
WITH AvgSalaryDept AS (
    SELECT department_id, AVG(salary) AS avg_salary
    FROM Employees
    GROUP BY department_id
)
SELECT *
FROM AvgSalaryDept;
```

```sql
--Use CTE to get employees and their department names
WITH EmpDept AS (
    SELECT e.emp_id, e.name, e.salary, d.dept_name
    FROM Employees e
    JOIN Departments d
    ON e.department_id = d.dept_id
)
SELECT *
FROM EmpDept;
```

```sql
--Find total order amount per employee using CTE
WITH TotalOrders AS (
    SELECT emp_id, SUM(order_amount) AS total_amount
    FROM Orders
    GROUP BY emp_id
)
SELECT *
FROM TotalOrders;
```

```sql
--Find employees whose salary is above department average using CTE
WITH DeptAvg AS (
    SELECT department_id, AVG(salary) AS avg_salary
    FROM Employees
    GROUP BY department_id
)
SELECT e.emp_id, e.name, e.salary, e.department_id
FROM Employees e
JOIN DeptAvg d
ON e.department_id = d.department_id
WHERE e.salary > d.avg_salary;
```
