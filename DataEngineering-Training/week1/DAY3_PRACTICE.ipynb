### 1.Assign a unique row number to all employees based on salary (highest first). 

```sql

select emp_id,emp_name,salary, row_number() OVER (ORDER BY salary DESC) AS row_num
FROM employees;
```

### 2.Assign row numbers to employees within each department based on salary descending. 

```sql

select emp_id,emp_name,department,salary, row_number() OVER (PARTITION BY department ORDER BY salary DESC) AS row_num
FROM employees;
```

### 3.Assign row numbers based on employee joining date (latest first) 

```sql
select *,row_number() OVER (ORDER BY join_date DESC) AS row_num
FROM employees;
```

### 4.Assign row numbers within each department based on earliest joining date

```sql
select *,row_number() OVER (Partition by department ORDER BY join_date DESC) AS row_num
FROM employees;
```

### 6.Assign row numbers to orders within each city based on order amount (highest first)

```sql
DROP TABLE orders;
CREATE TABLE orders (
    order_id INT,
    customer_id VARCHAR(10),
    customer_name VARCHAR(50),
    city VARCHAR(50),
    order_date DATE,
    order_amount DECIMAL(10,2)
);
INSERT INTO orders VALUES
(101, 'C001', 'Ravi',   'Chennai',  '2023-01-10', 250.50),
(102, 'C002', 'Anjali', 'Mumbai',   '2023-02-15', 300.00),
(103, 'C003', 'Kiran',  'Chennai',  '2023-03-05', 150.75),
(104, 'C004', 'Meena',  'Delhi',    '2023-01-25', 500.20),
(105, 'C005', 'Arjun',  'Mumbai',   '2023-03-20', 700.00),
(106, 'C006', 'Priya',  'Delhi',    '2023-02-05', 120.00);
```

```sql
SELECT 
    order_id,
    customer_id,
    city,
    order_amount,
    ROW_NUMBER() OVER (
        PARTITION BY city 
        ORDER BY order_amount DESC
    ) AS row_num
FROM orders;
```
