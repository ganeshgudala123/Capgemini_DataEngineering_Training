### Extracting Year,Month,Date

```sql
SELECT
YEAR(order_date),
MONTH(order_date),
DAY(order_date)
FROM orders;
```

```sql
SELECT
EXTRACT(YEAR FROM order_date),
EXTRACT(MONTH FROM order_date),
EXTRACT(DAY FROM order_date)
FROM orders;
```

### 8. WEEKDAY AND DAY OF WEEK

```sql
SELECT
WEEKDAY(order_date),
DAYOFWEEK(order_date)
FROM orders;
```

```sql
SELECT order_id, order_date
FROM orders
WHERE DAYOFWEEK(order_date) BETWEEN 2 AND 6;
```

```sql
SELECT order_date, ADD_months(order_date, 5 )
FROM orders;
```

```sql
SELECT 
    order_date, 
    ADD_MONTHS(order_date, 12 * 5) AS new_date
FROM orders;
```

```sql
select
    order_date,
    DATE_SUB(order_date, DAY(order_date) - 1) AS first_day_of_month
FROM orders;
```

### Setting DATE FORMAT

```sql
SELECT 
    DATE_FORMAT(order_date, 'dd-MM-yyyy') AS formatted_date
FROM orders;
```
