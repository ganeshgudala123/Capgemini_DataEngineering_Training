### 1 Use ROW_NUMBER() OVER (PARTITION BY student_id ORDER BY timestamp) to identify duplicates.

```sql
%python
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/workspace/default/capgemini_training/Capgemini Databricks VITB-2026 (1).csv")

df2 = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/workspace/default/capgemini_training/Task 1 (Responses).csv")

df3 = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/workspace/default/capgemini_training/Task 1 file 2.csv")
```

```sql
%python
df.createOrReplaceTempView("my_table")
df2.createOrReplaceTempView("my_table1")
df3.createOrReplaceTempView("my_table2")
```

```sql
SELECT *,
       ROW_NUMBER() OVER (
           PARTITION BY `Email address`
           ORDER BY Timestamp
       ) AS rn
FROM my_table2;
```

###  Keep first submission, mark others as duplicates

```sql
SELECT `Timestamp`,`Email address`,`  GitHub Username  `,
       CASE 
           WHEN ROW_NUMBER() OVER (
                    PARTITION BY `Email address` 
                    ORDER BY Timestamp
                ) = 1 
           THEN 'FIRST'
           ELSE 'DUPLICATE'
       END AS record_type
FROM my_table2;
```

### Count submissions per student

```sql
SELECT 
    `Email address`,
    COUNT(*) AS submission_count
FROM my_table2
GROUP BY `Email address`;
```
